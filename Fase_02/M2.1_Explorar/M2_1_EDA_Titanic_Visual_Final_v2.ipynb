```python
# --- M2.1_EDA_Titanic_Visual_Final.ipynb ---

# Objetivo: Consolidar la exploraci칩n inicial del dataset del Titanic con visualizaciones gr치ficas.
# A침adido: Histogramas, diagramas de caja y mapa de calor.

# 1. Importar librer칤as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci칩n b치sica para los gr치ficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# 2. Cargar el dataset del Titanic (usando la URL de GitHub)
github_csv_url = 'https://raw.githubusercontent.com/mortemj/curso-ml/main/Fase_02/M2.1_Explorar/titanic_train.csv'

try:
    df_titanic = pd.read_csv(github_csv_url)
    print("Dataset del Titanic cargado exitosamente para visualizaci칩n.")

    # --- 3.1: Primeras Vistas y Resumen ---
    print("\n--- Primeras 5 filas del DataFrame (df.head()) ---")
    print(df_titanic.head())

    print("\n--- Informaci칩n concisa del DataFrame (df.info()) ---")
    df_titanic.info()

    # --- 3.2: Visualizaciones Gr치ficas ---

    # Visualizaci칩n 1: Distribuci칩n de Supervivencia
    print("\n--- Gr치fico: Distribuci칩n de Supervivencia ---")
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Survived', data=df_titanic, palette='viridis')
    plt.title('Distribuci칩n de Supervivencia')
    plt.show()

    # Visualizaci칩n 2: Distribuci칩n por Sexo
    print("\n--- Gr치fico: Distribuci칩n por Sexo ---")
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Sex', data=df_titanic, palette='pastel')
    plt.title('Distribuci칩n de Pasajeros por Sexo')
    plt.show()

    # Visualizaci칩n 3: Distribuci칩n por Clase (Pclass)
    print("\n--- Gr치fico: Distribuci칩n por Clase de Pasajero (Pclass) ---")
    plt.figure(figsize=(7, 5))
    sns.countplot(x='Pclass', data=df_titanic, palette='magma')
    plt.title('Distribuci칩n de Pasajeros por Clase')
    plt.show()

    # Visualizaci칩n 4: Supervivencia por Sexo (usando 'hue')
    print("\n--- Gr치fico: Supervivencia por Sexo ---")
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Survived', hue='Sex', data=df_titanic, palette='coolwarm')
    plt.title('Supervivencia por Sexo')
    plt.show()

    # Visualizaci칩n 5: Histograma de la Edad
    print("\n--- Gr치fico: Histograma de la Edad ---")
    plt.figure(figsize=(8, 6))
    sns.histplot(df_titanic['Age'].dropna(), bins=30, kde=True, color='skyblue') # Elimina los NaN de la edad
    plt.title('Distribuci칩n de Edades')
    plt.xlabel('Edad')
    plt.ylabel('Frecuencia')
    plt.show()

    # Visualizaci칩n 6: Diagrama de Caja de la Edad por Clase
    print("\n--- Gr치fico: Diagrama de Caja de la Edad por Clase ---")
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Pclass', y='Age', data=df_titanic, palette='YlGnBu')
    plt.title('Distribuci칩n de Edades por Clase de Pasajero')
    plt.xlabel('Clase del Pasajero')
    plt.ylabel('Edad')
    plt.show()

    # Visualizaci칩n 7: Mapa de Calor de Correlaci칩n
    print("\n--- Gr치fico: Mapa de Calor de Correlaci칩n ---")
    plt.figure(figsize=(10, 8))
    sns.heatmap(df_titanic.corr(), annot=True, cmap='coolwarm', linewidths=.5) # 'annot=True' muestra los valores
    plt.title('Mapa de Calor de Correlaci칩n entre Variables Num칠ricas')
    plt.show()


except Exception as e:
    print(f"Error al cargar o visualizar el dataset: {e}")
    print("Aseg칰rate de que la URL en 'github_csv_url' es correcta y que el archivo CSV es accesible.")
```
