```python
# --- M2.1_EDA_Titanic_Visual_Final.ipynb ---

# Objetivo: Consolidar la exploración inicial del dataset del Titanic con visualizaciones gráficas.
# Añadido: Histogramas, diagramas de caja y mapa de calor.

# 1. Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración básica para los gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# 2. Cargar el dataset del Titanic (usando la URL de GitHub)
github_csv_url = 'https://raw.githubusercontent.com/mortemj/curso-ml/main/Fase_02/M2.1_Explorar/titanic_train.csv'

try:
    df_titanic = pd.read_csv(github_csv_url)
    print("Dataset del Titanic cargado exitosamente para visualización.")

    # --- 3.1: Primeras Vistas y Resumen ---
    print("\n--- Primeras 5 filas del DataFrame (df.head()) ---")
    print(df_titanic.head())

    print("\n--- Información concisa del DataFrame (df.info()) ---")
    df_titanic.info()

    # --- 3.2: Visualizaciones Gráficas ---

    # Visualización 1: Distribución de Supervivencia
    print("\n--- Gráfico: Distribución de Supervivencia ---")
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Survived', data=df_titanic, palette='viridis')
    plt.title('Distribución de Supervivencia')
    plt.show()

    # Visualización 2: Distribución por Sexo
    print("\n--- Gráfico: Distribución por Sexo ---")
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Sex', data=df_titanic, palette='pastel')
    plt.title('Distribución de Pasajeros por Sexo')
    plt.show()

    # Visualización 3: Distribución por Clase (Pclass)
    print("\n--- Gráfico: Distribución por Clase de Pasajero (Pclass) ---")
    plt.figure(figsize=(7, 5))
    sns.countplot(x='Pclass', data=df_titanic, palette='magma')
    plt.title('Distribución de Pasajeros por Clase')
    plt.show()

    # Visualización 4: Supervivencia por Sexo (usando 'hue')
    print("\n--- Gráfico: Supervivencia por Sexo ---")
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Survived', hue='Sex', data=df_titanic, palette='coolwarm')
    plt.title('Supervivencia por Sexo')
    plt.show()

    # Visualización 5: Histograma de la Edad
    print("\n--- Gráfico: Histograma de la Edad ---")
    plt.figure(figsize=(8, 6))
    sns.histplot(df_titanic['Age'].dropna(), bins=30, kde=True, color='skyblue') # Elimina los NaN de la edad
    plt.title('Distribución de Edades')
    plt.xlabel('Edad')
    plt.ylabel('Frecuencia')
    plt.show()

    # Visualización 6: Diagrama de Caja de la Edad por Clase
    print("\n--- Gráfico: Diagrama de Caja de la Edad por Clase ---")
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Pclass', y='Age', data=df_titanic, palette='YlGnBu')
    plt.title('Distribución de Edades por Clase de Pasajero')
    plt.xlabel('Clase del Pasajero')
    plt.ylabel('Edad')
    plt.show()

    # Visualización 7: Mapa de Calor de Correlación
    print("\n--- Gráfico: Mapa de Calor de Correlación ---")
    plt.figure(figsize=(10, 8))
    sns.heatmap(df_titanic.corr(), annot=True, cmap='coolwarm', linewidths=.5) # 'annot=True' muestra los valores
    plt.title('Mapa de Calor de Correlación entre Variables Numéricas')
    plt.show()


except Exception as e:
    print(f"Error al cargar o visualizar el dataset: {e}")
    print("Asegúrate de que la URL en 'github_csv_url' es correcta y que el archivo CSV es accesible.")
```
