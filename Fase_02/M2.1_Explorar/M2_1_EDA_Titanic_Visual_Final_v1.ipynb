```python
# --- M2.1_EDA_Titanic_Visual_Final_v1.ipynb ---

# Objetivo: Consolidar la exploración inicial del dataset del Titanic con visualizaciones gráficas.
# Añadido: Colores personalizados y uso del argumento 'hue'.

# 1. Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt # Para crear gráficos básicos
import seaborn as sns           # Para gráficos estadísticos más avanzados y estéticos

# Configuración básica para los gráficos (opcional, pero ayuda a que se vean mejor)
sns.set_style("whitegrid") # Establece un estilo de cuadrícula blanco para los gráficos
plt.rcParams['figure.figsize'] = (10, 6) # Establece el tamaño por defecto de los gráficos

# 2. Cargar el dataset del Titanic (usando la URL de GitHub que ya funciona)
# ¡Asegúrate de que esta URL es la CORRECTA y apunta a tu titanic_train.csv en GitHub!
github_csv_url = 'https://raw.githubusercontent.com/mortemj/curso-ml/main/Fase_02/M2.1_Explorar/titanic_train.csv'

try:
    df_titanic = pd.read_csv(github_csv_url)
    print("Dataset del Titanic cargado exitosamente para visualización.")

    # --- 3.1: Primeras Filas y Columnas ---
    print("\n--- Primeras 5 filas del DataFrame (df.head()) ---")
    print(df_titanic.head())

    print("\n--- Información concisa del DataFrame (df.info()) ---")
    df_titanic.info()

    # --- 3.2: Visualizaciones Gráficas con COLORES ---

    # Visualización 1: Distribución de Supervivencia con paleta de colores
    print("\n--- Gráfico: Distribución de Supervivencia (0=No, 1=Sí) con color ---")
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Survived', data=df_titanic, palette='viridis') # 'viridis' es una paleta secuencial
    plt.title('Distribución de Supervivencia')
    plt.xlabel('Supervivencia (0=No, 1=Sí)')
    plt.ylabel('Número de Pasajeros')
    plt.show()

    # Visualización 2: Distribución por Sexo con paleta de colores
    print("\n--- Gráfico: Distribución por Sexo con color ---")
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Sex', data=df_titanic, palette='pastel') # 'pastel' es una paleta discreta y suave
    plt.title('Distribución de Pasajeros por Sexo')
    plt.xlabel('Sexo')
    plt.ylabel('Número de Pasajeros')
    plt.show()

    # Visualización 3: Distribución por Clase (Pclass) con paleta de colores
    print("\n--- Gráfico: Distribución por Clase de Pasajero (Pclass) con color ---")
    plt.figure(figsize=(7, 5))
    sns.countplot(x='Pclass', data=df_titanic, palette='magma') # 'magma' es otra paleta secuencial
    plt.title('Distribución de Pasajeros por Clase')
    plt.xlabel('Clase del Pasajero')
    plt.ylabel('Número de Pasajeros')
    plt.show()

    # Visualización 4: Supervivencia por Sexo (usando 'hue' para añadir una dimensión de color)
    print("\n--- Gráfico: Supervivencia por Sexo (usando 'hue') ---")
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Survived', hue='Sex', data=df_titanic, palette='coolwarm') # 'coolwarm' para contrastar
    plt.title('Supervivencia por Sexo (0=No, 1=Sí)')
    plt.xlabel('Supervivencia')
    plt.ylabel('Número de Pasajeros')
    plt.legend(title='Sexo') # Añade una leyenda para 'Sex'
    plt.show()


except Exception as e:
    print(f"Error al cargar o visualizar el dataset: {e}")
    print("Asegúrate de que la URL en 'github_csv_url' es correcta y que el archivo CSV es accesible.")
```
