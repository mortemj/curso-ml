```python
# --- 02_exploracion_inicial_df.ipynb ---

# Objetivo: Realizar una primera inspección del dataset del Titanic.

# 1. Importar librerías necesarias
import pandas as pd

# 2. Cargar el dataset del Titanic (usando la URL de GitHub que ya funciona)
# ¡Asegúrate de que esta URL es la que ya te funciona y apunta a tu titanic_train.csv en GitHub!
github_csv_url = 'https://raw.githubusercontent.com/mortemj/curso-ml/main/Fase_02/M2.1_Explorar/titanic_train.csv'

try:
    df_titanic = pd.read_csv(github_csv_url)
    print("Dataset del Titanic cargado exitosamente para exploración.")

    # --- Primeras Vistas del Dataset ---

    print("\n--- 2.1: Primeras 5 filas del DataFrame (df.head()) ---")
    print("Esto nos muestra las primeras entradas y nos da una idea rápida de las columnas.")
    print(df_titanic.head())

    print("\n--- 2.2: Últimas 5 filas del DataFrame (df.tail()) ---")
    print("Útil para ver si hay datos atípicos o formatos diferentes al final.")
    print(df_titanic.tail())

    print("\n--- 2.3: Una muestra aleatoria del DataFrame (df.sample(5)) ---")
    print("Una buena forma de ver datos 'representativos' sin enfocarse en los extremos.")
    print(df_titanic.sample(5)) # Muestra 5 filas al azar

    print("\n--- 2.4: Dimensiones del DataFrame (df.shape) ---")
    print("Nos indica el número de filas (registros) y columnas (características).")
    print(f"Filas, Columnas: {df_titanic.shape}")

    print("\n--- 2.5: Nombres de las columnas (df.columns) ---")
    print("Lista todos los nombres de las características disponibles.")
    print(df_titanic.columns)

    print("\n--- 2.6: Información concisa del DataFrame (df.info()) ---")
    print("Fundamental para ver tipos de datos, conteo de valores no nulos y uso de memoria.")
    df_titanic.info() # No uses print() con .info() porque ya lo imprime por sí mismo

    print("\n--- 2.7: Estadísticas descriptivas de columnas numéricas (df.describe()) ---")
    print("Proporciona un resumen estadístico (media, desviación estándar, min, max, cuartiles) de las columnas numéricas.")
    print(df_titanic.describe())

    print("\n--- 2.8: Estadísticas descriptivas de columnas categóricas (df.describe(include='object')) ---")
    print("Útil para columnas de texto: cuenta valores únicos, valor más frecuente, etc.")
    print(df_titanic.describe(include='object'))

    print("\n--- 2.9: Conteo de valores únicos por columna (df.nunique()) ---")
    print("Nos dice cuántos valores diferentes hay en cada columna.")
    print(df_titanic.nunique())

except Exception as e:
    print(f"Error al cargar o inspeccionar el dataset: {e}")
    print("Asegúrate de que la URL en 'github_csv_url' es correcta y que el archivo CSV es accesible.")
```
